```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv("/content/3gpp_research_intents_2025-07-22.csv", engine='python', sep=',', on_bad_lines='warn')
```

```python
df
```

```python
df.shape
```

```python

df['clean_description'] = df['Description'].str.replace(r'\W+', ' ', regex=True).str.lower().str.strip()


label_encoder = LabelEncoder()
df['intent_label'] = label_encoder.fit_transform(df['Intent Type'])


text_col = 'clean_description'
categorical_cols = ['Priority', 'Network Slice', 'Location', 'Research Context', 'Compliance Standards']
numeric_cols = ['Technical Complexity']
target = 'intent_label'


X = df[[text_col] + categorical_cols + numeric_cols]
y = df[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, stratify=y, random_state=42)
```

```python
X_train_text = X_train[text_col].tolist()
X_test_text = X_test[text_col].tolist()
```

```python

vectorizer = TfidfVectorizer(max_features=1000)
X_train_vec = vectorizer.fit_transform(X_train_text)
X_test_vec = vectorizer.transform(X_test_text)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score
import numpy as np


X_train_dense = X_train_vec.toarray()
X_test_dense = X_test_vec.toarray()


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_dense)
X_test_scaled = scaler.transform(X_test_dense)



n_components = min(50, X_train_scaled.shape[0], X_train_scaled.shape[1])

pca = PCA(n_components=n_components)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

```python
logreg = LogisticRegression(penalty='l2', C=0.001, max_iter=200, solver='liblinear')
cv_scores = cross_val_score(logreg, X_train_pca, y_train, cv=5, scoring='f1_macro')
print("Cross-Validation F1 Scores:", cv_scores)
print("Mean F1 Score (LogReg + PCA):", np.mean(cv_scores))
```

```python
logreg.fit(X_train_pca, y_train)
logreg_preds = logreg.predict(X_test_pca)
print("Logistic Regression Test F1:", f1_score(y_test, logreg_preds, average='macro'))
```

```python
nb = MultinomialNB()

nb.fit(X_train_vec, y_train)
nb_preds = nb.predict(X_test_vec)
print("MultinomialNB Test F1:", f1_score(y_test, nb_preds, average='macro'))
```

```python

print("Logistic Regression Report:")
print(classification_report(y_test, logreg_preds, target_names=label_encoder.classes_))

print("\nNaive Bayes Report:")
print(classification_report(y_test, nb_preds, target_names=label_encoder.classes_))



```
